```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from numpy.random import randn, randint, uniform, sample
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.plot(x,y, 'go--', linewidth=1, markersize=5)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
#assingment 1
x=np.arange(0,10)
y=x*x
```

```python
#now we are going to plot the graph of the given variables
plt.plot(x,y, 'r+--', linewidth=1, markersize=10)
plt.xlabel('x-axis', color='orange')
plt.ylabel('y-axis', color='orange')
plt.title('2D-Diagram', color='blue')
```

```python
#now creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')



```
